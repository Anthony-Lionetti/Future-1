```python
import pandas as pd
# import numpy as np
# import plotly.express as px
# import matplotlib.pyplot as plt
import requests
```

**Some mildly useful functions**

```python
def url(endpoint:str, session_key:str) -> str:
    path = f'https://api.openf1.org/v1/{endpoint}?session_key={session_key}'
    return path

def res_to_df(res:requests.Response) -> pd.DataFrame:
    if res.status_code != 200:
        print(f'Problem making request. Status code: {res.status_code}')
        return pd.DataFrame()
    
    data = res.json()
    return pd.json_normalize(data)
```

## Looking at the Drivers Endpoint

```python
endpoint = "drivers"
session_key = 'latest'
res = requests.get(url(endpoint, session_key))
df = res_to_df(res)
```
